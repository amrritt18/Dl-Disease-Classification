# Ensure Annotation

```python
def get_product (x:int,y:int) -> int: 
    return x*y
```

```python
get_product(x=2,y=3)
```

```python
get_product(x=2,y='3')
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product (x:int,y:int) -> int: 
    return x*y
```

```python
get_product(x=2,y=3)
```

```python
get_product(x=2,y='3')
```

# Download Data from Google Drive

```python
import gdown
```

```python
url = "https://drive.google.com/file/d/18JXqBCrGluH3KhKnOcm5A1rGmTw361Vy/view?usp=sharing"
```

```python
url.split("/")
```

```python
file_id = url.split("/")[-2]
file_id
```

```python
prefix = "https://drive.google.com/uc?export=download&id="
gdown.download(prefix + file_id, "kidney-CT-Scan-data.zip", quiet=False)
```
